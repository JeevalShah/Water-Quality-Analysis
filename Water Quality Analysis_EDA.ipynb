## Visualising the distribution of data

```r
# Box Plot of Each Column (excluding Potability)
numeric_columns <- names(data)[sapply(data, is.numeric)]  # Selecting numeric columns
```

```r
plots <- list()

# Loop through each numeric column and create boxplots
for (col in numeric_columns) {
  if (col != "Potability") {
    p <- ggplot(data, aes_string(y = col, fill = "factor(Potability)")) +
      geom_boxplot(na.rm = TRUE) +  # Removes rows with non-finite values
      labs(title = paste("Boxplot of", col), y = col, fill = "Potability") +
      theme_minimal() +
      theme(
        axis.text.x = element_text(size = 8),  # Adjusting axis text size
        axis.text.y = element_text(size = 8),
        plot.title = element_text(size = 12)  # Reducing title size
      )
    
    # Add the plot to the list
    plots[[col]] <- p
  }
}

# Use grid.arrange to display all plots with adjusted width and height
do.call(grid.arrange, c(plots, ncol = 3, nrow = 3))
```

```r
plots
```

### Analysis of distribution of data
The boxplots provide a visual representation of the distribution of various water quality parameters for both potable and non-potable water. Each boxplot shows the median, quartiles, and potential outliers in the data for the respective water quality features. By examining these boxplots, we can develop an intuitive understanding of the relationships between these variables and water potability.

## Graphs

```r
# Set color scheme for Potability
colors <- c("0" = "red", "1" = "green")

# Create scatterplots for each numeric variable against 'ph'
plots <- list()  # Initialize an empty list to store plots

# Loop through all columns except 'Potability' (last column)
for (col in names(data)[1:9]) {  
  p <- ggplot(data, aes_string(x = "ph", y = col, color = "factor(Potability)")) +
    geom_point(alpha = 0.6, shape = 16) +  # Use shape 16 for dots
    scale_color_manual(values = colors) +
    labs(title = paste(col, "vs pH"), x = "pH", y = col, color = "Potability") +
    theme_minimal()
  
  plots[[col]] <- p  # Add each plot to the list
}

# Use grid.arrange to display all 9 plots in a 3x3 grid
do.call(grid.arrange, c(plots, ncol = 3, nrow = 3))
```

```r
plots
```

```r
# Create bar charts for each numeric column with binned values and colored by Potability
plots <- list()  # Initialize an empty list to store plots

# Loop through all columns except 'Potability'
for (col in names(data)[1:9]) {
  # Create bins for the column (10 bins)
  data[[paste0(col, "_bins")]] <- cut(data[[col]], breaks = 10, include.lowest = TRUE)
  
  # Create a bar chart for the binned data
  p <- ggplot(data, aes_string(x = paste0(col, "_bins"), fill = "factor(Potability)")) +
    geom_bar(position = "dodge") +
    scale_fill_manual(values = colors) +
    labs(title = paste("Distribution of", col), x = col, y = "Count", fill = "Potability") +
    theme_minimal() +
      theme(
        axis.text.x = element_text(size = 6, angle = 45, hjust = 1), # Adjusting axis text size
        axis.text.y = element_text(size = 8),
        plot.title = element_text(size = 10)
      )
  plots[[col]] <- p  # Add each plot to the list
}
```

## Skewness & Kurtosis

```r
data <- read.csv("C:/Users/G-One/OneDrive/Desktop/Probability & Statistics/Project/water_potability.csv")

cleaned_data <- na.omit(data)

# Calculate skewness and kurtosis
skewness_kurtosis <- cleaned_data %>%
  summarise(across(everything(), list(
    skewness = ~ skew(.), 
    kurtosis = ~ kurtosis(.)
  )))
```

```r
print(skewness_kurtosis)
```

### Analysis of Skewness and Kurtosis

The skewness and kurtosis values provide insights into the shape and distribution of the data for each variable. Here's a breakdown of the results:

#### 1. **pH**
- **Skewness**: 0.0489 (Approximately symmetric)
- **Kurtosis**: 3.6186 (Leptokurtic)
  - The positive skewness indicates a slight right tail, but it is close to zero, suggesting symmetry. The kurtosis value above 3 indicates a leptokurtic distribution, meaning it has heavier tails and a sharper peak than a normal distribution.

#### 2. **Hardness**
- **Skewness**: -0.0851 (Approximately symmetric)
- **Kurtosis**: 3.5229 (Leptokurtic)
  - The skewness is slightly negative, suggesting a very slight left tail. The kurtosis value also indicates a leptokurtic distribution, implying it has heavier tails and a sharper peak.

#### 3. **Solids**
- **Skewness**: 0.5950 (Moderately positively skewed)
- **Kurtosis**: 3.3412 (Leptokurtic)
  - The positive skewness indicates a longer right tail. The kurtosis value suggests a leptokurtic distribution, implying it has heavier tails and a sharper peak compared to a normal distribution.

#### 4. **Chloramines**
- **Skewness**: 0.0130 (Approximately symmetric)
- **Kurtosis**: 3.5472 (Leptokurtic)
  - The skewness indicates a symmetric distribution, while the kurtosis shows a leptokurtic shape, suggesting a peakier distribution with heavier tails.

#### 5. **Sulfate**
- **Skewness**: -0.0465 (Approximately symmetric)
- **Kurtosis**: 3.7824 (Leptokurtic)
  - Similar to chloramines, the skewness indicates symmetry. The high kurtosis suggests a leptokurtic distribution, indicating more extreme values.

#### 6. **Conductivity**
- **Skewness**: 0.2665 (Slightly positively skewed)
- **Kurtosis**: 2.7581 (Platykurtic)
  - The positive skewness indicates a slight right tail. However, the kurtosis value being less than 3 signifies a platykurtic distribution, meaning it has lighter tails and a flatter peak compared to a normal distribution.

#### 7. **Organic Carbon**
- **Skewness**: -0.0200 (Approximately symmetric)
- **Kurtosis**: 3.0294 (Mesokurtic)
  - The skewness indicates a symmetric distribution. The kurtosis close to 3 indicates a mesokurtic distribution, which is similar to a normal distribution in terms of tail behavior and peak.

#### 8. **Trihalomethanes**
- **Skewness**: -0.0513 (Approximately symmetric)
- **Kurtosis**: 3.2211 (Leptokurtic)
  - The skewness indicates symmetry, while the kurtosis value suggests a leptokurtic distribution, indicating heavier tails and a sharper peak.

#### 9. **Turbidity**
- **Skewness**: -0.0330 (Approximately symmetric)
- **Kurtosis**: 2.9478 (Platykurtic)
  - The skewness indicates a symmetric distribution, and the kurtosis less than 3 suggests a platykurtic shape, with lighter tails.

#### 10. **Potability**
- **Skewness**: 0.3940 (Moderately positively skewed)
- **Kurtosis**: 1.1555 (Platykurtic)
  - The positive skewness indicates a longer right tail, and the kurtosis value being less than 3 signifies a platykurtic distribution, indicating a flatter peak and lighter tails compared to normal distribution.

### Summary:
- **Leptokurtic**: pH, Hardness, Solids, Chloramines, Sulfate, Trihalomethanes
- **Platykurtic**: Conductivity, Turbidity, Potability
- **Mesokurtic**: Organic Carbon

Overall, most variables display a leptokurtic distribution, which suggests that they are prone to producing outliers. Conductivity, Turbidity, and Potability. However, have platykurtic distributions, indicating fewer extreme values compared to a normal distribution.
